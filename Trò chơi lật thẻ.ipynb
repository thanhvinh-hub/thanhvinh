```python
import tkinter as tk
from tkinter import messagebox
import random

class MemoryGame:
    def __init__(self, root):
        self.root = root
        self.root.title("Memory Card - Nguyên tố hóa học")

        # Dữ liệu nguyên tố
        self.elements = [
            ("Hydrogen", "H"), ("Helium", "He"), ("Carbon", "C"),
            ("Nitrogen", "N"), ("Oxygen", "O"), ("Sodium", "Na"),
            ("Calcium", "Ca"), ("Iron", "Fe"), ("Sulfur", "S"), ("Chlorine", "Cl")
        ]

        self.setup_game()

    def setup_game(self):
        # Tạo danh sách thẻ (Tên + Kí hiệu)
        self.cards_data = []
        for name, symbol in self.elements:
            self.cards_data.append({'text': name, 'pair': symbol})
            self.cards_data.append({'text': symbol, 'pair': name})

        random.shuffle(self.cards_data)

        self.buttons = []
        self.first_card = None
        self.second_card = None
        self.lock_board = False
        self.matches = 0

        # Giao diện điểm số
        self.score_label = tk.Label(self.root, text="Điểm: 0", font=("Arial", 14))
        self.score_label.pack(pady=10)

        # Khung chứa các quân bài
        self.frame = tk.Frame(self.root)
        self.frame.pack(padx=20, pady=20)

        for i in range(len(self.cards_data)):
            btn = tk.Button(self.frame, text="?", width=10, height=4,
                            command=lambda i=i: self.flip_card(i),
                            font=("Arial", 10, "bold"), bg="#ffcc70")
            btn.grid(row=i//5, column=i%5, padx=5, pady=5)
            self.buttons.append(btn)

    def flip_card(self, index):
        if self.lock_board or self.buttons[index]['text'] != "?":
            return

        # Hiển thị nội dung thẻ
        self.buttons[index].config(text=self.cards_data[index]['text'], bg="white")

        if not self.first_card:
            self.first_card = index
        else:
            self.second_card = index
            self.check_match()

    def check_match(self):
        self.lock_board = True
        txt1 = self.cards_data[self.first_card]['text']
        pair2 = self.cards_data[self.second_card]['pair']

        if txt1 == pair2:
            # Nếu khớp
            self.root.after(500, self.hide_cards)
        else:
            # Nếu không khớp, lật lại sau 800ms
            self.root.after(800, self.reset_cards)

    def hide_cards(self):
        self.buttons[self.first_card].grid_forget()
        self.buttons[self.second_card].grid_forget()
        self.matches += 1
        self.score_label.config(text=f"Điểm: {self.matches}")

        if self.matches == len(self.elements):
            messagebox.showinfo("Chúc mừng", "🎉 Bạn đã thuộc hết các nguyên tố!")

        self.reset_turn()

    def reset_cards(self):
        self.buttons[self.first_card].config(text="?", bg="#ffcc70")
        self.buttons[self.second_card].config(text="?", bg="#ffcc70")
        self.reset_turn()

    def reset_turn(self):
        self.first_card = None
        self.second_card = None
        self.lock_board = False

if __name__ == "__main__":
    root = tk.Tk()
    game = MemoryGame(root)
    root.mainloop()
```
